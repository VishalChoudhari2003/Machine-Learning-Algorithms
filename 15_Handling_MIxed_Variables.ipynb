```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/new titatnic.csv')
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
fig=df['number'].value_counts().plot(kind='bar')
fig.set_title('Passenge Travelling with')
```

```python
#Extracty numerical part
df['number_numerical']=pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
#Extract categorical part
df['number_categorical']=np.where(df['number_numerical'].isnull(),df['number'],np.nan)
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['cabin_num ']=df['Cabin'].str.extract('(\d+)')
df['cabin_cat']=df['Cabin'].str[0]
```

```python
df.head()
```

```python
#Extract the last bit of Ticket
df['ticket_num']=df['Ticket'].apply(lambda s:s.split()[-1])
df['ticket_num']=pd.to_numeric(df['ticket_num'],errors='coerce',downcast='integer')
```

```python
#Extract Categorical
df['ticket_cat']=np.where(df['ticket_num'].isnull(),df['Ticket'],np.nan)
```
