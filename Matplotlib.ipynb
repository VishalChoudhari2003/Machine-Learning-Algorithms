##Matplotlib


```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
#sin wave
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show() #cosine wave
```

```python
#Adding title
plt.plot(x,y)
plt.title('sin wave')
plt.show()
```

```python
plt.plot(x,z)
plt.title('cosine wave')
plt.show()
```

```python
plt.plot(x,y,label='sin wave')
plt.plot(x,z,label='cosine wave')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define data
languages = ['English', 'French', 'German']
students = [23, 17, 35]

# Create figure and axis
fig1 = plt.figure()
ax = fig1.add_axes([0, 0, 1, 1])

# Corrected pie chart syntax
ax.pie(students, labels=languages, autopct='%1.1f%%')

# Display the plot
plt.show()
```
