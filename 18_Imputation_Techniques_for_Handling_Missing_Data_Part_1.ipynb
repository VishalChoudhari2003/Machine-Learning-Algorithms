##Arbitary Imputation

```python
import pandas as pd

# Sample dataset
data = {'Age': [22, 38, None, 26, 35], 'Fare': [7.25, 71.28, 8.05, None, 53.10], 'Survived': [1, 1, 0, 1, 0]}
df = pd.DataFrame(data)

# Apply arbitrary imputation
df['Age'].fillna(-1, inplace=True)  # Replace missing Age with -1
df['Fare'].fillna(999, inplace=True)  # Replace missing Fare with 999

# Display updated dataset
print(df)
```

```python
from sklearn.impute import SimpleImputer

# Define arbitrary value imputer
imputer = SimpleImputer(strategy="constant", fill_value=-1)  # Arbitrary value -1

# Apply to Age column
df['Age'] = imputer.fit_transform(df[['Age']])
```
