Imbalanved dataset:
 A dataset with an unequal class distribution

```python
 import pandas as pd
 import numpy as np
```

```python
#loading the dataset
credit_card_data=pd.read_csv('credit_data.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.shape
```

```python
credit_card_data['Class'].value_counts()
```

This is Highly Imbalance dataset

0-->legit transaction
1-->fraudulent transaction

```python
#Separating the ledgit and fraudulent data
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under-Sampling

Building a sample dataset containing similar distribution of L and F transactrions

```python
legit_sample=legit.sample(n=417)
```

```python
print(legit_sample.shape)
```

Concatenate two dataframes

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```
