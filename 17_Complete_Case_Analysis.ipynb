```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/data_science_job.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().mean()*100
```

```python
cols=[var for var in df.columns if df[var].isnull().mean()<0.05 and df[var].isnull().mean()>0]
cols
```

```python
df[cols].sample(5)
```

```python
len(df[cols].dropna())/len(df)
```

```python
new_df=df[cols].dropna()
df.shape,new_df.shape
```

```python
new_df.hist(bins=50,density=True,figsize=(20,15))
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)

#original data
df['training_hours'].hist(bins=50,density=True,alpha=0.5,ax=ax,color='red')

#new data
new_df['training_hours'].hist(bins=50,density=True,alpha=0.5,ax=ax,color='blue')
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)

#original data
df['city_development_index'].hist(bins=50,density=True,alpha=0.5,ax=ax,color='red')

#new data
new_df['city_development_index'].hist(bins=50,density=True,alpha=0.5,ax=ax,color='blue')
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)

#original data
df['experience'].hist(bins=50,density=True,alpha=0.5,ax=ax,color='red')

#new data
new_df['experience'].hist(bins=50,density=True,alpha=0.5,ax=ax,color='blue')
```

```python
temp=pd.concat([
    #percentage of observation per category original data
    df['enrolled_university'].value_counts()/len(df),

    #percentage of observation per category cca data
    new_df['enrolled_university'].value_counts()/len(new_df)
],axis=1)
temp.columns=['Original','CCA']
temp
```

```python
temp=pd.concat([
    #percentage of observation per category original data
    df['education_level'].value_counts()/len(df),

    #percentage of observation per category cca data
    new_df['education_level'].value_counts()/len(new_df)
],axis=1)
temp.columns=['Original','CCA']
temp
```
